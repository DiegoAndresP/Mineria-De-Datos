# Clustering y Clasificacion de Datos

 El codigo utiliza el algoritmo de clustering K-Means para agrupar los registros de un conjunto de datos en función de sus valores de ventas globales promedio por año. Luego, genera una gráfica de dispersión para visualizar los grupos.


- Los datos se escalan utilizando la técnica de estandarización para asegurarse de que todas las características tengan la misma escala.
- Se crea un modelo de K-Means con 3 clusters y se ajusta a los datos escalados.
- Cada registro se asigna a un cluster en función de su similitud con los centroides de los clusters.
- Se agrega una columna "Cluster" al DataFrame original para almacenar las etiquetas de los clusters.
- La gráfica de dispersión muestra los registros en el eje x correspondiente a las ventas globales promedio y en el eje y a una línea horizontal. Cada punto se colorea según su cluster asignado.
- La gráfica permite visualizar cómo se agrupan los registros en base a sus ventas globales promedio. Los puntos del mismo color representan registros similares en términos de ventas globales promedio.

En resumen, el código utiliza K-Means para agrupar los datos en base a las ventas globales promedio y muestra la distribución de los clusters en la gráfica de dispersión. Esto ayuda a identificar patrones o segmentos en los datos según sus niveles de ventas globales promedio.

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Cargar datos
df = pd.read_csv('/home/slayer24/WorkSpace/MINERIA_DE_DATOS/DataSets/Ds_estadistica_desc/promedio_ventas_gl_año.csv')

# Seleccionar la columna de interés
X = df[['GLOBAL_VENTAS']].values

# Escalar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Definir el modelo K-Means y ajustarlo a los datos
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)

# Asignar una etiqueta a cada registro
labels = kmeans.labels_

# Agregar las etiquetas al DataFrame
df['Cluster'] = labels

# Visualizar los grupos
colors = ['red', 'green', 'blue']
for i in range(3):
    plt.scatter(X[labels == i], np.zeros_like(X[labels == i]), color=colors[i])
plt.xlabel('Promedio Ventas Globales')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Cargar datos
df = pd.read_csv('/home/slayer24/WorkSpace/MINERIA_DE_DATOS/DataSets/Ds_estadistica_desc/Cantidad_vd_año.csv')

# Eliminar filas con valores nulos
df.dropna(inplace=True)

# Seleccionar las columnas de interés
X = df[['CANTIDAD', 'AÑO']].values

# Normalizar los datos
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Elegir el número de clusters
num_clusters = 3

# Entrenar el modelo
kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(X)

# Obtener los centroides de los clusters
centroids = kmeans.cluster_centers_

# Obtener las etiquetas de los clusters para cada punto de datos
labels = kmeans.labels_

# Visualizar los clusters y los centroides
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200, linewidths=3, color='r')
plt.xlabel('Cantidad de ventas')
plt.ylabel('Año')
plt.title('Clustering con K-means')
plt.show()
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Cargar el conjunto de datos
df = pd.read_csv('/home/slayer24/WorkSpace/MINERIA_DE_DATOS/DataSets/Ds_estadistica_desc/juegos_mayor_ventas_GLOBAL_VENTAS.csv')

# Seleccionar las columnas de interés
df_filtrado = df[['NOMBRE', 'GLOBAL_VENTAS']]

# Preprocesamiento de datos
X = np.array(df_filtrado['GLOBAL_VENTAS']).reshape(-1, 1)

# Definir el número de clusters
k = 3

# Entrenar el modelo K-Means
kmeans = KMeans(n_clusters=k, random_state=0).fit(X)

# Obtener las etiquetas y los centroides
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Visualización de los resultados
plt.scatter(X, np.zeros_like(X), c=labels)
plt.scatter(centroids, np.zeros_like(centroids), marker='x', s=200, linewidths=3, color='r')
plt.xlabel('Mayor Ventas Globales(millones)')
plt.show()
```

En este código se realiza clustering utilizando el algoritmo K-Means para agrupar los datos de ventas globales de videojuegos por plataforma. A continuación, se explica el funcionamiento:

1. Se carga el conjunto de datos desde un archivo CSV y se seleccionan las columnas de interés, en este caso, "PLATAFORMA" y "GLOBAL_VENTAS".

2. La columna "PLATAFORMA" se codifica utilizando One-Hot Encoding para convertir las diferentes plataformas en variables binarias.

3. Se escala el conjunto de datos utilizando StandardScaler para asegurar que todas las variables tengan la misma escala.

4. Se define el número de clusters deseado, en este caso, n_clusters = 3.

5. Se entrena el modelo de K-Means utilizando los datos escalados.

6. Se asignan etiquetas de cluster a cada fila del dataframe original.

7. Se imprimen los resultados del clustering para cada cluster, incluyendo la cantidad de filas en cada cluster, las plataformas más comunes en cada cluster, y el promedio de ventas globales en cada cluster.

8. Se grafican los clusters, donde cada punto representa un videojuego. Los puntos se colorean según su etiqueta de cluster y se muestran los centroides de los clusters.

Se agrupan los videojuegos según sus ventas globales y la plataforma en la que se lanzaron. Luego, muestra los resultados del clustering y una gráfica que representa los clusters y los centroides. Esto permite identificar patrones y segmentos en los datos de ventas globales de videojuegos por plataforma.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder


# Cargar datos
df = pd.read_csv('/home/slayer24/WorkSpace/MINERIA_DE_DATOS/DataSets/vgsales_limp.csv')

# Seleccionar las columnas de interés
df_filtrado = df[['PLATAFORMA', 'GLOBAL_VENTAS']]

# Codificar la columna "PLATAFORMA" usando One-Hot Encoding
encoder = OneHotEncoder()
plataformas_encoded = encoder.fit_transform(df_filtrado[['PLATAFORMA']]).toarray()
plataformas_df = pd.DataFrame(plataformas_encoded, columns=encoder.get_feature_names_out(['PLATAFORMA']))
df_filtrado = pd.concat([plataformas_df, df_filtrado[['GLOBAL_VENTAS']]], axis=1)

# Escalar los datos
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_filtrado)

# Realizar clustering con k-means
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(df_scaled)

# Asignar etiquetas de cluster a cada fila
df_filtrado['cluster'] = kmeans.labels_

# Imprimir los resultados
for i in range(n_clusters):
    cluster = df_filtrado[df_filtrado['cluster'] == i]
    print(f"Cluster {i}:")
    print(f"Cantidad de filas: {len(cluster)}")
    print(f"Plataformas más comunes: {cluster.iloc[:, :7].sum().sort_values(ascending=False)[:5].index.tolist()}")
    print(f"Promedio de ventas globales: {cluster['GLOBAL_VENTAS'].mean()}")
    print('\n')

# Graficar los clusters
colors = ['blue', 'red', 'green']
plt.figure(figsize=(10,6))
for i in range(n_clusters):
    cluster = df_filtrado[df_filtrado['cluster'] == i]
    plt.scatter(cluster.iloc[:,0], cluster.iloc[:,1], s=50, c=colors[i], label=f"Cluster {i}")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, marker='s', c='black', alpha=0.7, label='Centroides')
plt.xlabel('Plataformas')
plt.ylabel('Ventas Globales')
plt.legend()
plt.show()
```
