```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

cantidad_vd_año= pd.read_csv('/home/slayer24/WorkSpace/MINERIA_DE_DATOS/DataSets/Ds_estadistica_desc/Cantidad_vd_año.csv')

# Crear un objeto de regresión lineal
reg = LinearRegression()

# Definir las variables independiente y dependiente
X = cantidad_vd_año[['AÑO']]
y = cantidad_vd_año['CANTIDAD']

# Ajustar la regresión lineal
reg.fit(X, y)

# Imprimir la pendiente y el intercepto de la recta
print("Pendiente: ", reg.coef_[0])
print("Intercepto: ", reg.intercept_)

# Graficar los puntos de datos y la recta ajustada
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red')
plt.xlabel('AÑO')
plt.ylabel('CANTIDAD')

plt.savefig('/home/slayer24/WorkSpace/MINERIA_DE_DATOS/Lineal_Models/Graficas_ML/cantidad_vd_año.png')
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Cargamos los datos
prediccion_ventas_NA = pd.read_csv('/home/slayer24/WorkSpace/MINERIA_DE_DATOS/DataSets/vgsales_limp.csv')

# Elegimos una variable numérica para predecir las ventas globales (en este caso, las ventas de Norteamérica)
x = prediccion_ventas_NA['N.A_VENTAS'].values.reshape(-1,1)
y = prediccion_ventas_NA['GLOBAL_VENTAS'].values.reshape(-1,1)

# Creamos un modelo de regresión lineal
reg = LinearRegression()
reg.fit(x, y)

# Predecimos las ventas globales para un valor de ventas en Norteamérica
ventas_norteamerica = [[10]]  # valor a predecir
prediccion = reg.predict(ventas_norteamerica)

# Imprimimos la pendiente (coeficiente) y la intercepción del modelo
print('Pendiente:', reg.coef_[0][0])
print('Intercepción:', reg.intercept_[0])

# Graficamos los puntos y la recta ajustada por el modelo
plt.scatter(x, y, color='blue')
plt.plot(x, reg.predict(x), color='red', linewidth=2)
plt.title('Prediccion Ventas Norteamerica')
plt.xlabel('Ventas en Norteamérica')
plt.ylabel('Ventas globales')

plt.savefig('/home/slayer24/WorkSpace/MINERIA_DE_DATOS/Lineal_Models/Graficas_ML/predicciones_ventas_NA.png')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Carga el dataframe
df = pd.read_csv('/home/slayer24/WorkSpace/MINERIA_DE_DATOS/DataSets/vgsales_limp.csv')

# Selecciona las columnas numéricas
X = df[['N.A_VENTAS', 'EUR_VENTAS', 'JP_VENTAS', 'OTRAS_VENTAS']].values
y = df['GLOBAL_VENTAS'].values

# Divide los datos en conjunto de entrenamiento y prueba
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Entrena el modelo de regresión polinómica
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X_train)
pol_reg = LinearRegression()
pol_reg.fit(X_poly, y_train)

# Realiza una predicción utilizando el conjunto de prueba
y_pred = pol_reg.predict(poly_reg.transform(X_test))

# Visualiza los resultados
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].scatter(X_train[:, 0], y_train, color='red')
axs[0, 0].plot(X_train[:, 0], pol_reg.predict(X_poly), color='blue')
axs[0, 0].set_title('Norteamerica')
axs[0, 0].set_xlabel('Ventas Norteamerica')
axs[0, 0].set_ylabel('Ventas globales')

axs[0, 1].scatter(X_train[:, 1], y_train, color='red')
axs[0, 1].plot(X_train[:, 1], pol_reg.predict(X_poly), color='blue')
axs[0, 1].set_title('Europa')
axs[0, 1].set_xlabel('Ventas Europa')
axs[0, 1].set_ylabel('Ventas globales')

axs[1, 0].scatter(X_train[:, 2], y_train, color='red')
axs[1, 0].plot(X_train[:, 2], pol_reg.predict(X_poly), color='blue')
axs[1, 0].set_title('Asia')
axs[1, 0].set_xlabel('Ventas Asia')
axs[1, 0].set_ylabel('Ventas globales')

axs[1, 1].scatter(X_train[:, 3], y_train, color='red')
axs[1, 1].plot(X_train[:, 3], pol_reg.predict(X_poly), color='blue')
axs[1, 1].set_title('Resto del Mundo')
axs[1, 1].set_xlabel('Ventas Resto Mundo')
axs[1, 1].set_ylabel('Ventas globales')

plt.savefig('/home/slayer24/WorkSpace/MINERIA_DE_DATOS/Lineal_Models/Graficas_ML/Regresion_Polinomial.png')
plt.show()
```
