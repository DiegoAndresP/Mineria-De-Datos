```python
import pandas as pd
import matplotlib.pyplot as plt


ventas_genero = pd.read_csv('/home/slayer24/WorkSpace/MINERIA_DE_DATOS/DataSets/Ds_estadistica_desc/ventas_genero.csv')

ventas_promedio = ventas_genero.groupby('GENERO')['GLOBAL_VENTAS'].mean()


ax = ventas_promedio.plot(kind='bar', color='orange', alpha=0.8, figsize=(8,6))


ax.set_title('Ventas por género')
ax.set_xlabel('Género')
ax.set_ylabel('Ventas (en millones)')

plt.savefig('/home/slayer24/WorkSpace/MINERIA_DE_DATOS/Visualization_Data/Graficas_png/Ventas_Genero.png', dpi=100, bbox_inches='tight')
plt.show()


```
