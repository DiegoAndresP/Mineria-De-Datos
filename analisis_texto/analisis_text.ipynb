```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Cargar los datos en un dataframe
df = pd.read_csv('/home/slayer24/Workspace/MINERIA_DE_DATOS/DataSets/vgsales_limp.csv')

# Preprocesar los datos si es necesario
text = ' '.join(df['NOMBRE'].astype(str))

# Crear y visualizar la nube de palabras
wordcloud = WordCloud(width=800, height=400, max_font_size=150, random_state=42).generate(text)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
